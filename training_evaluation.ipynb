BERT Loss with Attention based on Attention Pairs

```python
import json
import matplotlib.pyplot as plt

with open("loss_log_Attention_Pairs.json", "r") as f:
    loss_data = json.load(f)

epochs = [d["epoch"] for d in loss_data]
train_losses = [d["train_loss"] for d in loss_data]
val_losses = [d["val_loss"] for d in loss_data]

plt.plot(epochs, train_losses, label="Train Loss")
plt.plot(epochs, val_losses, label="Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Validation Loss")
plt.legend()
plt.grid(True)
plt.show()
```
